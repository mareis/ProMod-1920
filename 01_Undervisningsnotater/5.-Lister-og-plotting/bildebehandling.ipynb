```python
np.shape(img)[0:2]
```

```python
gs = np.zeros(shape=np.shape(img)[0:2])
for x in range(np.shape(img)[1]):
    for y in range(np.shape(img)[0]):
        m = np.mean(img[y,x])
        gs[y,x] = m

plt.imshow(gs, cmap='gray')
```

```python
gray_image = img.dot([0.07, 0.72, 0.21])
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
plt.imshow(np.abs(gray_image-255), cmap='gray')
```

```python
plt.imshow(255-gray_image, cmap="gray")
```

```python
gray_image[gray_image>150] = 255
plt.imshow(gray_image, cmap='gray')
```
